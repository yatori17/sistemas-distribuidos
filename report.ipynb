```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
report = pd.read_csv('report.csv')
```

```python
report
```

```python
report_1worker = report[report.num_workers == 1]
report_5workers = report[report.num_workers == 5]
report_15workers = report[report.num_workers == 15]
report_20workers = report[report.num_workers == 20]
```

```python
fig, ax = plt.subplots()
ax.plot(report_1worker.num_files_searched, report_1worker.execution_time, label='1 worker')
ax.plot(report_5workers.num_files_searched, report_5workers.execution_time, label='5 workers')
ax.plot(report_15workers.num_files_searched, report_15workers.execution_time, label='15 workers')
ax.plot(report_20workers.num_files_searched, report_20workers.execution_time, label='20 workers')
plt.legend()
plt.show()
```

```python
report2 = pd.read_csv('report2.csv')
```

```python
report2.head()
```

```python
num_workers = report2.num_workers.unique()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
for i in num_workers:
    ax.plot(report2[report2.num_workers == i].num_files_searched, report2[report2.num_workers == i].execution_time, label='{} worker(s)'.format(i))

plt.ylabel('Execution Time (s)')
plt.xlabel('Number of Files Searched')
ax.set_ylim(0,0.6)
ax.set_xlim(0,55)
plt.legend(bbox_to_anchor =(0.65, 1.25))
plt.legend()
plt.show()
```
